# Translating Principles into Practice

While AI ethics principles tell us "what" we should and should not be doing, they generally don't tell us "how" to do it.

"How" usually entails organizational governance - some of which will be process-oriented, some of which will be model-oriented. The "how" is summarized in an AI Ethics Framework.

What is an AI Ethics Framework?
- a document that breaks each AI ethics principle into one or more specific and measureable actions to be completed by employees
- checklist/series of questions/SOP

How does an AI Ethics Framework integrate into AI Ethics Principles?
- High-level Principle <br>
    - Sub-principles <br>
        - Framework Question (or questions)<br>

Development Process
1) high-level principles and corresponding sub-principles are developed (by an interdiscplinary team) and approved <br>
2) framework questions are developed to cover each aspect of every sub-principles (by an interdiscplinary team that includes technical representatives - ideally data scientist who will be implementing the framework) and approved <br>
3) the framework is integrating into an existing model development process <br>
4) the model development team completes the framework at the initiation of each new AI project AND revisits it when changes are made to the model(s) AND on a pre-defined calendar  
            

What are the some key learnings to making an AI Ethics Framework work?
-  integration into existing modeling framework (CRISP-DM, Data Science Methodology, MLOps)
-  integration of existing privacy, cybersecurity, and risk-management procedures

Examples of existing modeling frameworks: IBM's CRISP-DM and Data Science Methodology (retrieved from: https://developer.ibm.com/articles/introduction-watson-studio/)

# Diversity, Non-discrimination and Fairness Framework
Positive Predictive Value (PPV) was the fairness metric chosen by the organization in the example above to implement 1 of 5 principles of Diversity, Equity, Fairness and Non-discrimination. PPV is just one of over 25 fairness metrics used in practice.


In technical ML practice Fairness = non-discrimination = anti-discrimination
- although in reality these things are not equal


There are 3 main categories of fairness metrics: <br>
1) Classification Parity <br>
- Fairness is achieved when some pre-defined measure of classification error is equal across protected groups <br>
- Includes various metrics calculated from a confusion matrix including error rate balance and statistical parity <br>
- Statistical parity (also referred to as equal acceptance rate) attempts to ensure that at a given decision threshold, the proportion of individuals with predicted probabilities above the threshold is the same across protected groups <br>
- Error rate balance is achieved if at a given threshold, the false positive and false negatives rates are equal across protected groups <br>
- Some researchers acknowledge the practical challenge of perfect balance and instead measure the difference in error rates

2) Calibration <br>
- A well-calibrated (fair) model produces outcomes that are independent of protected attributes, conditional on some predeteremined risk score <br>
- Example - United Health App found to be discriminating against black individuals (in the US) using a calibration fairness metric: https://www.healthcarefinancenews.com/news/study-finds-racial-bias-optum-algorithm

3) Anti-classification <br>
- Algorithmic decision-making models that do not consider protected attributes (read: exclude protected variables, like gender and age), with stronger definitions suggesting that even proxy variables for the protected attributes be excluded <br>
- This is a naive metric that has been found to significantly increase discrimination against protected groups


A few challenges with fairness metrics:
1) Several fairness metrics are not achievable at the same time<br>
2) Simply optimizing for fairness can be highly detrimental to model accuracy (and ultimately profitability)<br>
3) Many fairness metrics were originally developed in the context of the judicial system and race - essentially focusing on equity (not necessarily equality)<br>
4) Different fairness metrics require different types of data<br>

# A note on preserving unequal default rates

Equality= a model with equality in mind would be equally good at predicting one gender vs. another - and existing unequal default rates would be preserved

Equity= a model with equity in mind would change the default rates to equalize them between two gender, and existing unequal default rates would NOT be preserved 

In the context of credit scoring we generally want any measure of discrimination to adjust for unequal base default rates between protected groups (in this case, genders). If instead we used a measure of discrimination that created equity (the same # of defaults for women as men) then we would be unfairly discriminating against the more creditworthy group (in this case women).

# Using pre-processing methods to reduce discrimination

In our original dataset there were four predictions from pre-processed models:

Pre-processing methods to reduce discrimination (and their predictions)

HTM2 = predictions from a hyperparameter-tuned (HT) model, without gender (M2) <br>
- this model includes gender as a feature during the hyperparameter tuning, but then excludes it as a feature when training the full model <br>
- this method requires certain ML model types and platforms which allow for customization (hyperparameter tuning)<br>
- for example, this pre-processing method is not feasible in AI-as-a-service tools like DataRobot<br>

DSM2 = predictions from a model where the majority population is downsampled (DS) to match the majority<br>
- prior to training the model, the dataset is resampled so that the count of the majority population matches the 
- various resampling strategies can be used depending on performance (e.g., random, stratified random) <br>

USM2 = predictions from a model where the minority population is upsampled (US) to match the majority<br>
- similar to down-sampling, except the majority population is upsampled (e.g., women are upsampled to match the count of men)<br>
- again, various resampling strategies can be used (e.g., SMOTE, random, stratefied random)<br>

PGPM2 = predictions from a probabilistic gender proxy model (PGP) which first predicts gender, then uses the prediction in place of actual gender for new applicants <br>
- two models are run in this pre-processing method
- first, a model is run to predict gender of past borrowers
- then the applicant data is run through this "gender prediction" model to predict the gender of new applicants
- the predicted gender is then included in the dataset for applicants and a second model is trained to predict default on this dataset

Each of these methods varies in its ability to reduce discrimination depending on the threshold selected by the modeling team to determine default. However, you can look at the results across a range of thresholds (5 - 30%) and make some general conclusions.

A reminder that we looked at one measure of discrimination (PPV), but there are several other fairness (or discriminatino) metrics, many of which are discussed at IBM's AI Fairness 360 website: https://aif360.readthedocs.io/en/stable/modules/sklearn.html#module-aif360.sklearn.metrics

# Linking things back to an existing ML process

We will conclude the tutorial by revisiting how to integrate ML fairness (and other AI ethics principles) into the broader organizational operations.

The key to ensuring the framework items are achieved (and therefore the AI ethics principles are upheld) is to integrate the framework questions into an existing process. 



Within the data science team's documentation, each specific framework question (e.g."Will you develop a measure of algorithmic fairness with cleargo/no-goKPIs, specific to each AI model, knowing thatno single definition of fairness will apply equally well to different applications of AI?") needs to be assigned to one or more steps within the existing process.

For example, perhaps the fairness framework question we addressed in the example is should be first addressed in the "Analytic Approach" step when the team is decided upon the model type (and therefore approach fairness metric to use).
